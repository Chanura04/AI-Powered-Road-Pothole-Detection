```python

import shutil
folder_A = "C:/Projects Datasets/RoadDamageDetection/train/Czech/images"
folder_B = "C:/Projects Datasets/RoadDamageDetection/train/Czech/annotations/xmls"

output_save="C:\Projects Datasets\RoadDamageDetection\\preprocessed_01"

files_A = set(os.listdir(folder_A))
files_B = set(os.listdir(folder_B))

image_to_add=image_files[:2000]
anno_to_add=anno_500


images_out = os.path.join(output_save, "images")
annos_out = os.path.join(output_save, "annotations")
os.makedirs(images_out, exist_ok=True)
os.makedirs(annos_out, exist_ok=True)


image_to_copy = [f for f in image_to_add if f in files_A]
anno_to_copy  = [f for f in anno_to_add  if f in files_B]


# --- Copy images ---
copied_images = 0
for filename in image_to_copy:
    src = os.path.join(folder_A, filename)
    dst = os.path.join(images_out, filename)
    try:
        shutil.copy2(src, dst)  # preserves metadata
        copied_images += 1
    except Exception as e:
        print(f"[ERROR] Failed to copy image '{filename}': {e}")

# --- Copy annotations ---
copied_annos = 0
for filename in anno_to_copy:
    src = os.path.join(folder_B, filename)
    dst = os.path.join(annos_out, filename)
    try:
        shutil.copy2(src, dst)
        copied_annos += 1
    except Exception as e:
        print(f"[ERROR] Failed to copy annotation '{filename}': {e}")

```
