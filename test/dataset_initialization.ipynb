```python

import os
import re
import shutil
from pathlib import Path

# ==== CONFIGURE THESE ====
NUMBERS_SOURCE = final_anno_set_numbers
ANNO_DIR = "C:\Projects Datasets\RoadDamageDetection\\preprocessed_01\\annotations"
OUTPUT_DIR = "C:\Projects Datasets\RoadDamageDetection\handled_imbalance_processed_data\\annotations"       # Where to copy matched files
OUTPUT_CSV = "matched_list.csv"     # Report of matches

# Allowed image extensions
ANNO_EXTS = {".xml"}

def load_numbers():
    return NUMBERS_SOURCE

def match_filename(name_stem, targets):
    # name_stem example: "crez_001"
    try:
        number = int(name_stem.split('_')[1])  # "001"  1
    except (IndexError, ValueError):
        return None

    if number in targets:
        return number

    return None

def main():
    numbers = load_numbers()
    images_dir = Path(ANNO_DIR)
    out_dir = Path(OUTPUT_DIR)
    out_dir.mkdir(parents=True, exist_ok=True)

    matches = []  # list of (number, filename)

    for file in images_dir.iterdir():
        if not file.is_file():
            continue
        if file.suffix.lower() not in ANNO_EXTS:
            continue

        stem = file.stem  # filename without extension
        stem=stem.split('.')[0]
        matched = match_filename(stem, numbers)
        if matched:
            # Copy to output folder, keep original name
            dest = out_dir / file.name
            # If name collision, append a counter
            if dest.exists():
                base, ext = dest.stem, dest.suffix
                i = 1
                while True:
                    candidate = out_dir / f"{base}_{i}{ext}"
                    if not candidate.exists():
                        dest = candidate
                        break
                    i += 1
            shutil.copy2(file, dest)
            matches.append((matched, file.name))


    print(f"Done. Matched {len(matches)} image(s).")
    print(f"- Copied to: {out_dir.resolve()}")
    print(f"- Report:    {Path(OUTPUT_CSV).resolve()}")

if __name__ == "__main__":
    main()
```
